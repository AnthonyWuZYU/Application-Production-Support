```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Rating_curve = pd.read_csv ('/Users/shg096/Desktop/Rating_Crve_05BB001.csv')
measurment   = pd.read_csv ('/Users/shg096/Desktop/measurment_05BB001.csv')

print(Rating_curve)
print(measurment)
```

```python
# plot

plt.figure(figsize=(20,10))

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.grid()

plt.scatter(Rating_curve['stage'], Rating_curve['dicharge'], label='Rating Curve')

plt.scatter(measurment['stage'], measurment['dicharge'], label = 'Measurement')

plt.legend(loc= 2,prop={'size': 30})

plt.xlim(1, 4.5) 

plt.title ('Bow at Banff; 05BB001', size= 30)

plt.savefig ('/Users/shg096/Desktop/Rating_Curve_vs_Measurement_Bow_Banff_05BB001.png')
```
